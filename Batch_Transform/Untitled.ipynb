# Scikit-Learn

```python
import sagemaker

session = sagemaker.Session(default_bucket = "demo-output-bucket", default_bucket_prefix = "Batch_Transform/") 
# default_bucket_prefix is not available in older versions of SageMaker
role = sagemaker.get_execution_role()
```

```python
image_url = "720646828776.dkr.ecr.ap-south-1.amazonaws.com/sagemaker-scikit-learn:1.0-1-cpu-py3"
model_data_url = "s3://demo-output-bucket/Training_Pipeline_Output/2023-09-08T07:35:39.843Z/HPTuningOutputs/Decision_Tree/gxvc0dt117zh-hptuning-cQVa6D6efJ-001-3a4cf63d/output/model.tar.gz"
instance_type = "ml.c5.xlarge"
```

```python
from sagemaker.model import Model
model = Model(
    entry_point = "Evaluation.py",
    source_dir = "SKLearn", 
    role = role,
    image_uri = image_url,
    model_data = model_data_url
)
```

```python
predictor = model.deploy(initial_instance_count=1, instance_type=instance_type, endpoint_name = "churn-endpoint-11111")
```

```python
transformer = sagemaker.transformer.Transformer(
    model_name = "sagemaker-scikit-learn-2023-10-12-10-15-37-184",
    instance_count = 1,
    instance_type = "ml.c5.xlarge",
    sagemaker_session = session
)
```

```python
transformer.transform("s3://demo-inputs-bucket/Batch_Data/Batch_Data.csv"
                      # , content_type = 'text/csv'
                     )
```
